```python

file_path = f'/content/drive/Shareddrives/QRLSSP 2024/COVID/Python Code/Mobility/Processed Daily CSVs/*.csv'
```

```python
import pandas as pd
import glob

# Step 1: Read the main CSV file
file_path_pattern = file_path
csv_files = glob.glob(file_path_pattern)

data_frames = []
cnt=0
for file in csv_files:
    df = pd.read_csv(file)
    cnt=cnt+1
    data_frames.append(df)

filtered_df = pd.concat(data_frames, ignore_index=True)
filtered_df.head()
print(cnt)
```

```python
# Group by 'category_o' and 'category_d' and calculate sums
totals = filtered_df.groupby(['category_o', 'category_d']).agg({
    'visitor_flows': 'sum',
    'pop_flows': 'sum'
}).reset_index()

print(totals)
```

```python
directory = "/content/drive/Shareddrives/QRLSSP 2024/COVID/Python Code/Mobility/Final CSV of 45 days/"
filename = f"{directory}total_mobility.csv"
totals.to_csv(filename, index=False)
```
