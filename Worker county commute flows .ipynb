```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/drive/Shareddrives/QRLSSP 2024/COVID/Useful CSVs/2016-2020-worker-county-commute-flows.csv',skiprows=7,encoding='ISO-8859-1')
df.head(15)
```

```python
df['County FIPS Code'] = df['County FIPS Code'].astype(str).str.split('.').str[0].str.zfill(3)
df['County FIPS Code.1'] = df['County FIPS Code.1'].astype(str).str.split('.').str[0].str.zfill(3)

df['State FIPS Code.1'] = df['State FIPS Code.1'].replace([np.inf, -np.inf], np.nan).fillna(0)

# Convert 'State FIPS Code.1' from float to integer
df['State FIPS Code.1'] = df['State FIPS Code.1'].astype(int)
# Display the updated DataFrame
print(df.head())
```

```python
# Merge State FIPS Code with County FIPS Code
df['Residence FIPS Code'] = df['State FIPS Code'].astype(str).str.zfill(1) + df['County FIPS Code']
df['Workplace FIPS Code'] = df['State FIPS Code.1'].astype(str).str.zfill(1) + df['County FIPS Code.1']
```

```python
df.head()
```

```python
filtered_df=df[['Residence FIPS Code','Workplace FIPS Code','Workers in Commuting Flow','Margin of Error']]
```

```python
filtered_df.head()
```

```python
filtered_df = filtered_df[filtered_df['Residence FIPS Code'] != filtered_df['Workplace FIPS Code']]
```

```python
filtered_df.head()
```

```python
fips_df=pd.read_csv('/content/drive/Shareddrives/QRLSSP 2024/COVID/Useful CSVs/fips_categories.csv')
fips_df['fips'] = fips_df['fips'].astype(str)
fips_df
```

```python
filtered_df = filtered_df.merge(fips_df, left_on='Residence FIPS Code', right_on='fips', how='left')
filtered_df.rename(columns={'category': 'category_residence'}, inplace=True)
filtered_df.drop(columns=['fips', 'name'], inplace=True)

# Merge for category_workplace
filtered_df = filtered_df.merge(fips_df, left_on='Workplace FIPS Code', right_on='fips', how='left')
filtered_df.rename(columns={'category': 'category_workplace'}, inplace=True)
filtered_df.drop(columns=['fips', 'name'], inplace=True)
```

```python
filtered_df.head()
```

```python
filtered_df = filtered_df[filtered_df['category_residence'] != filtered_df['category_workplace']]
filtered_df.head()
```

```python
filtered_df['Workers in Commuting Flow'] = pd.to_numeric(filtered_df['Workers in Commuting Flow'], errors='coerce')
filtered_df['Margin of Error'] = pd.to_numeric(filtered_df['Margin of Error'], errors='coerce')
```

```python
# Group by categories and aggregate sums
totals = filtered_df.groupby(['category_residence', 'category_workplace']).agg({
    'Workers in Commuting Flow': 'sum',
    'Margin of Error': 'sum'
}).reset_index()

# Set pandas display options for float formatting
pd.options.display.float_format = '{:,.0f}'.format

# Display the totals DataFrame
print(totals)
```

```python
pivot_table = totals.pivot(index='category_residence', columns='category_workplace', values='Workers in Commuting Flow').fillna(0)

print(pivot_table)
```
